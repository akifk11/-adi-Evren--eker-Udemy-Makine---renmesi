```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv("veriler.csv")
veriler
```

```python
ulke = veriler.iloc[:,0:1].values
ulke
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
ulke[:,0] = le.fit_transform(veriler.iloc[:,0])
#burada processing yaptım yani sınıflandırma yaptım.
ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke).toarray()
ulke
```

```python
c = veriler.iloc[:,-1:].values
c
le = preprocessing.LabelEncoder()
c[:,-1] = le.fit_transform(veriler.iloc[:,-1])
ohe = preprocessing.OneHotEncoder()
c = ohe.fit_transform(c).toarray()
c
```

```python
veriler1 = veriler.iloc[:,1:4]
veriler1
```

```python
ulke = pd.DataFrame(data=ulke, index = range(22), columns=["tr","fr","us"])
```

```python
import statsmodels.api as sm
```

```python
#burada değişkenlerden hangisi p value değerini daha çok etkiliyorsa onu çıkartıcaz bunun için bu işlemleri yapıyoruz bu backward elminationdır
```

```python
X = np.append(arr = np.ones((22,1)).astype(int), values = veri, axis = 1)
#en başa birlerden oluşan bir kolon ekledi.
X
```

```python
X_l = veri.iloc[:,[0,1,2,3,4,5]].values
X_l
```

```python
X_l = np.array(X_l,dtype=float)
X_l
```

```python
#OLS istatistiksel değerleri çıkarmaya yarar
model = sm.OLS(boy,X_l).fit()
# boy ile bağımsız değişken olan data arasında ilişki kuruyoruz
```

```python
model.summary()
#bütün amaç OlS regression sonuçlarını elde etmek için bu raporu almak çok önemli bizim için şuan önemli olan p değerleri
#burada amacımız backward algoritması adımları gereği en yüksek p değerini elemek
```

```python
# şuan 4 index teki elemanı çıkartıp yeniden hesaplama yapıyoruz
X_l = veri.iloc[:,[0,1,2,3,5]].values
X_l = np.array(X_l,dtype = float)
model = sm.OLS(boy,X_l).fit()
model.summary()
```

```python
#daha iyi bir sonuç için aynı işlemi tekrarlıyoruz
X_l = veri.iloc[:,[0,1,2,3]].values
X_l = np.array(X_l,dtype = float)
model = sm.OLS(boy,X_l).fit()
model.summary()
#statsmodel ile modelin istatistiksel değerlerine ulaştık
```
